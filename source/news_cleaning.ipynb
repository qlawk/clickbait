```python
import pandas as pd
import numpy as np
```

```python
news = pd.read_csv('../data/news.csv')
news.head()
```

```python
# checking for patterns, trends in data
news.info()
```

```python
news.describe(exclude=np.number).T
```

```python
# dropping irrelvant columns
news = news.drop(columns=['article_id','source_id','author','description','url','url_to_image','published_at','content','category','article'])
```

```python
# renaming columns
news = news.rename(columns={'title':'headline','source_name':'source'})
```

```python
news.count()
```

```python
# drop non-english headlines
from langdetect import detect

news['language'] = news['headline'].apply(detect)
news = news[news['language'] == 'en']
```

```python
news = news.drop(columns='language')
```
